```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ramp_1 = pd.read_csv("more_data_v2.csv")
ramp_2 = pd.read_csv("response.csv")

step_1 = pd.read_csv("step.csv")
step_2 = pd.read_csv("step_v2.csv")
```

```python
plt.plot(step_1.query("Tempo < 2.5")["Tempo"], step_1.query("Tempo < 2.5")["Referencia"], c="r", label="Referência")
plt.plot(step_1.query("Tempo < 2.5")["Tempo"], step_1.query("Tempo < 2.5")["Posicao"], c="g", label="Saída")
plt.plot(step_1.query("Tempo < 2.5")["Tempo"], step_1.query("Tempo < 2.5")["Erro"], c="b", label="Erro")
plt.grid()
plt.title("Resposta ao degrau")
plt.ylabel("Tensão [V]")
plt.xlabel("Tempo [s]")
plt.legend()
plt.show()
```
